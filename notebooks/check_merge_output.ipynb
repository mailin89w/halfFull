# Sanity Check after Merge

```python
from pathlib import Path
import pandas as pd

file_path = Path("..") / "data" / "processed" / "nhanes_merged_adults.csv"
print(file_path.resolve())

df = pd.read_csv(file_path, low_memory=False)
print(df.shape)
print(df["SEQN"].nunique())
```

```python
print(df["SEQN"].nunique())
print(df["SEQN"].duplicated().sum())
print(df["SEQN"].isna().sum())

print
```

```python
missing = df.isna().sum().sort_values(ascending=False)
missing.head(30)
```
